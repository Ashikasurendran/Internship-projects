## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
#check missing values
df.isnull().sum()
```

```python
#check duplicates
df.duplicated().sum()
```

```python
#remove cancelled orders
#Cancelled invoices start with "C"
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
#Remove negative quantity
df = df[df['Quantity'] > 0]
```

```python
#remove zero price
df = df[df['UnitPrice'] > 0]
```

```python
#Convert Data Types
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%d-%m-%Y %H:%M')
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
#Create total price column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

#### Chart - **8 RFM  Analysis**

```python
# Chart - 8 visualization code
snapshot_date = df['InvoiceDate'].max() + dt.timedelta(days=1)
```

```python
#calculate RFM
rfm=df.groupby('CustomerID').agg({'InvoiceDate': lambda x: (snapshot_date - x.max()).days,'Invoi    eNo': 'count','TotalPrice': 'sum'})
rfm.rename(columns={'InvoiceDate': 'Recency', 'Invoi    eNo': 'Frequency', 'TotalPrice': 'Monetary'}, inplace=True)
rfm.head()
```

#### **RFM Distribution Visualization**

```python
#Recency Distribution
rfm['Recency'].plot(kind='hist',bins=50)
plt.title('Recency Distribution')
plt.show
```

```python
#Frequency Distribution
rfm['Frequency'].plot(kind='hist',bins=50)
plt.title('Frequency Distribution')
plt.show
```

```python
#Monetary Distribution
rfm['Monetary'].plot(kind='hist',bins=50)
plt.title('Monetary Distribution')
plt.show
```

**RFM Segmentation**

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'], 4, labels=[1, 2, 3, 4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

rfm['RFM-Score'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm.head()
```
