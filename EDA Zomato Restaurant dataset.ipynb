## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1  Top Cuisines

````python
top_cuisines = merged_df['Cuisines'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index)
plt.title("Top 10 Cuisines")
plt.xlabel("Count")
plt.ylabel("Cuisines")
plt.show()
````

##### 1. What is/are the insight(s) found from the chart?

Bar chart is used to compare the frequency of different cuisines. It helps identify the most popular cuisines.

#### Chart - 2 Cost Distribution

````python
# Chart - 2 visualization code
plt.figure(figsize=(10,5))
sns.histplot(merged_df['Cost'], bins=20)

plt.title("Cost Distribution")
plt.xlabel("Cost")
plt.ylabel("Count")

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
````

##### 1. What is/are the insight(s) found from the chart?

Histogram helps understand how restaurant costs are distributed.

#### Chart - 3 Cost vs Rating 
(Bivariate Analysis (Relationship between 2 variables))

````python
# Chart - 3 visualization code
plt.xlabel("Cost")
plt.ylabel("Rating")

plt.xticks(rotation=45)
plt.tight_layout()
plt.figure(figsize=(8,5))
sns.scatterplot(x=merged_df['Cost'], y=merged_df['Rating'])
plt.title("Cost vs Rating")
plt.show()
````

##### 1. What is/are the insight(s) found from the chart?

Scatter plot shows the relationship between cost and rating.

#### Chart - 6   Top Restaurants (based on number of reviews or entries)

````python
# Chart - 6 visualization code
top_restaurants = merged_df['Name'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_restaurants.values, y=top_restaurants.index)

plt.title("Top 10 Restaurants")
plt.xlabel("Number of Entries / Reviews")
plt.ylabel("Restaurant Name")

plt.show()
````

#### Chart - 7  Top Restaurants based on Rating

````python
#Convert Rating to numeric
merged_df['Rating'] = pd.to_numeric(merged_df['Rating'], errors='coerce')
````

````python
#remove missing rates
merged_df = merged_df.dropna(subset=['Rating'])
````

````python
# Chart - 7 visualization code
top_restaurants = merged_df.groupby('Name')['Rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_restaurants.values, y=top_restaurants.index)

plt.title("Top 10 Restaurants by Average Rating")
plt.xlabel("Average Rating")
plt.ylabel("Restaurant Name")

plt.show()
````

##### 1.  what is are the isight(s) found from the chart?
The rating column was converted to numeric format to enable aggregation operations like mean. Invalid or non-numeric values were handled by converting them into null values and removing them to ensure accurate analysis.
